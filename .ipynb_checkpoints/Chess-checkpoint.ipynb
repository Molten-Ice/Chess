```python
def letters_to_index(letter_number):
    number_2 = int(letter_number[1]) - 1
    number_1 = ord(str(letter_number[0]).upper())-64 -1
    if number_1 < 0 or number_1 > 7 or number_2 <0 or number_2 >7:
        raise ValueError('Index out of range') 
    
    return [number_1, number_2]
class Board:
    def __init__(self):
        c1 = "black"
        c2 = "white"
        self.board = [[Rook(c1), Knight(c1), Bishop(c1), Queen(c1), King(c1), Bishop(c1), Knight(c1), Rook(c1)],
                      [Pawn(c1), Pawn(c1)  , Pawn(c1)  , Pawn(c1) , Pawn(c1), Pawn(c1)  , Pawn(c1)  , Pawn(c1)],
                      [blank   , blank     , blank     , blank    , blank   , blank     , blank     , blank   ],
                      [blank   , blank     , blank     , blank    , blank   , blank     , blank     , blank   ],
                      [blank   , blank     , blank     , blank    , blank   , blank     , blank     , blank   ],
                      [blank   , blank     , blank     , blank    , blank   , blank     , blank     , blank   ],
                      [Pawn(c2), Pawn(c2)  , Pawn(c2)  , Pawn(c2) , Pawn(c2), Pawn(c2)  , Pawn(c2)  , Pawn(c2)],
                      [Rook(c2), Knight(c2), Bishop(c2), Queen(c2), King(c2), Bishop(c2), Knight(c2), Rook(c2)]]
        
    def move(self):
        raw_input = input("Enter piece to move and where to move:")
        #raw_input = "A2 A3"
        valid_input = True
        try:
            piece_to_move = letters_to_index(raw_input[:2])
            spot_to_move_to = letters_to_index(raw_input[3:5])
        except Exception as e:
            print("Invalid input")
            print("Exception:", e)
            valid_input = False
            
        if valid_input and self.board[piece_to_move[1]][piece_to_move[0]] == blank:
            print("No piece is selected coordinates")
            valid_input = False
            
        if valid_input:
            self.board[piece_to_move[1]][piece_to_move[0]].move(piece_to_move, spot_to_move_to, self)
 
            
    def __repr__(self):
        return_string = ""
        
        return_string+="  "
        for i in range(0,8):
            return_string+=  str(chr(65+i)) +" ｜"
            
        return_string += "\n"
        
        top_value = 1
        for row in self.board:
            return_string+=str(top_value) +" "
            top_value+=1
            for element in row:
                return_string+= str(element) + "｜"
            return_string+= "\n"
        return return_string
```

```python

blank = "  "
class Piece:
    def __init__(self, colour):
        self.colour = colour
        self.character = "None"
        
    def __repr__(self):
        return self.character
    
class Pawn(Piece):
    def __init__(self, colour):
        Piece.__init__(self, colour)
        if self.colour == "white":
            self.character = "♙"
        else:
            self.character = "♟︎"
            
    def move(self, piece_to_move, spot_to_move_to, board_object):
        vector_to_move = [spot_to_move_to[0]-piece_to_move[0], spot_to_move_to[1]-piece_to_move[1]]
        if self.colour == "black":
            multiplying_factor = 1 # Downwards
        else:
            multiplying_factor = -1 # Upwards
        vector_to_move_scaled = [vector_to_move[0]*1,vector_to_move[1]*multiplying_factor]
        if board_object.board[spot_to_move_to[1]][spot_to_move_to[0]] == blank:
            if vector_to_move_scaled == [0, 1]:
                #single forward
                board_object.board[spot_to_move_to[1]][spot_to_move_to[0]] =\
                board_object.board[piece_to_move[1]][piece_to_move[0]]
                board_object.board[piece_to_move[1]][piece_to_move[0]] = blank
            elif vector_to_move_scaled == [0, 2] and (piece_to_move[1] == 1 or piece_to_move[1] == 6):
                #double forward
                board_object.board[spot_to_move_to[1]][spot_to_move_to[0]] =\
                board_object.board[piece_to_move[1]][piece_to_move[0]]
                board_object.board[piece_to_move[1]][piece_to_move[0]] = blank
            else:

                print("Invalid move selection", spot_to_move_to)
        elif board_object.board[spot_to_move_to[1]][spot_to_move_to[0]].colour != \
                board_object.board[piece_to_move[1]][piece_to_move[0]].colour:
#             print("Foo", vector_to_move, vector_to_move_scaled)
            if vector_to_move_scaled == [-1, 1] or vector_to_move_scaled == [1, 1]:
                #Take piece

                board_object.board[spot_to_move_to[1]][spot_to_move_to[0]] =\
                board_object.board[piece_to_move[1]][piece_to_move[0]]
                board_object.board[piece_to_move[1]][piece_to_move[0]] = blank
            else:
                print("Invalid move selection")
        
class Knight(Piece):
    #Horse
    def __init__(self, colour):
        Piece.__init__(self, colour)
        if self.colour == "white":
            self.character = "♘"
        else:
            self.character = "♞"
            
    def move(self, piece_to_move, spot_to_move_to, board_object):
        vector_to_move = [spot_to_move_to[0]-piece_to_move[0], spot_to_move_to[1]-piece_to_move[1]]
        allowed_vectors = [[1,2],[-1,2],[-2,1],[-1,-2], [2,1],[-2,1],[2,-1],[-2,-1]]
        if board_object.board[spot_to_move_to[1]][spot_to_move_to[0]] == blank or \
        board_object.board[spot_to_move_to[1]][spot_to_move_to[0]].colour != self.colour:
            if vector_to_move in allowed_vectors:
                board_object.board[spot_to_move_to[1]][spot_to_move_to[0]] =\
                board_object.board[piece_to_move[1]][piece_to_move[0]]
                board_object.board[piece_to_move[1]][piece_to_move[0]] = blank    
            else:
                print("Invalid move selection")
        else:
            print("Location occupied")
        
class Bishop(Piece):
    #Diagonal
    def __init__(self, colour):
        Piece.__init__(self, colour)
        if self.colour == "white":
            self.character = "♗"
        else:
            self.character = "♝"
        
class Rook(Piece):
    #Castle
    def __init__(self, colour):
        Piece.__init__(self, colour)
        if self.colour == "white":
            self.character = "♖"
        else:
            self.character = "♜"
            
class Queen(Piece):
    def __init__(self, colour):
        Piece.__init__(self, colour)
        if self.colour == "white":
            self.character = "♕"
        else:
            self.character = "♛"
            
class King(Piece):
    def __init__(self, colour):
        Piece.__init__(self, colour)
        if self.colour == "white":
            self.character = "♔"
        else:
            self.character = "♚"
            
```
